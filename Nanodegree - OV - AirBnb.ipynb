## 1. Least requested area(s) in Seattle

## 2. Is there a correlation between expensive stays and good reviews?

```python
# get the data
expensive = lis[['neighbourhood', 'number_of_reviews', 'review_scores_rating', 'price']]
```

```python
# remove listings with less than 15 reviews to prevent outliers
expensive = expensive[expensive.number_of_reviews >=15]
```

```python
expensive['price'].sort_values(ascending=False)
```

```python
# remove the dollar sign and cast to number format to be able to calculate the correlation
expensive['price'] = expensive['price'].str.replace('$', '')
expensive['price'] = expensive['price'].astype(float)
```

```python
expensive['price'].corr(expensive['review_scores_rating'])
```

There seems to be a small positive correlation.
